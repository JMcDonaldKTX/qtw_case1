```python
from sklearnex import patch_sklearn
patch_sklearn()

import pandas as pd
from sklearn.linear_model import LassoCV, RidgeCV, LinearRegression
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler
```

##### Build linear regression models with L1 and L2 regularization to predict the critical
##### temperature as closely as possible.  Include which variable carries the most importance.

```python
df_train = pd.read_csv('train.csv')
df_unique_m = pd.read_csv('unique_m.csv')
```

```python
#concatenate the two dfs on axis=1 and drop the critical_temp from the first
#ASSUMES THAT THE ROW # IS IDENTITY AS NO OTHER COLUMN IS DEFINED AS SUCH
df = pd.concat([df_train.drop(['critical_temp'], axis=1), df_unique_m], axis=1, join='inner')
df
```

#### Prepare my data
I'm setting up a train/test split with 15% held back for testing performance.

```python
X = df.drop(columns=['critical_temp', 'material'])
y = df['critical_temp']

Xscaler = StandardScaler()
X = Xscaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.15, random_state=2023)
```

#### Lasso and Ridge CV
Starting with basice lasso and ridge regression with cross validation.

```python
lasso = LassoCV(cv=5, random_state=2023).fit(X_train, y_train)
ridge = RidgeCV(cv=5).fit(X_train, y_train)
```

```python
l1_score = lasso.score(X_test, y_test)
l2_score = ridge.score(X_test, y_test)

print("L1 Score: ", l1_score)
print("L2 Score: ", l2_score)
```

#### Grid Search CV
Next I'll try grid search approaches to determine the appropriate metrics for each and come up with a best performing model


```python
param_grid_linreg = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100],'penalty':['l1', 'l2']}
reg = LinearRegression()
grid_search = GridSearchCV(reg, param_grid_linreg, cv=5)
grid_search.fit(X_train, y_train)

print("Best parameters: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
```
